```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
all_df = pd.read_csv('train.csv')

# Split the data into training (70%) and testing (30%) sets
train_df, test_df = train_test_split(all_df, test_size=0.3, random_state=42)

# Display the shapes of the two datasets
print("Train DataFrame shape:", train_df.shape)
print("Test DataFrame shape:", test_df.shape)
```

```python
train_df.corr()
```

```python
correlation_with_saleprice = train_df.corr()['SalePrice']

# Sort correlations in descending order
highest_correlations_with_saleprice = correlation_with_saleprice.abs().sort_values(ascending=False)

# Display the top correlations with SalePrice
print("Top Correlations with SalePrice:")
print(highest_correlations_with_saleprice.head(20))
```

```python
sns.scatterplot(data=train_df, x="OverallQual", y="SalePrice")
```

```python
#Average slopes in OverallQual vs SalePrice

def getAverageSlope(df):
    average_Slopes = []
    for i in range(len(df['SalePrice'])):
        slope = df.iloc[i]['SalePrice'] / df.iloc[i]['OverallQual']
        average_Slopes.append(slope)
    return sum(average_Slopes) / len(average_Slopes) 

averageSlope = getAverageSlope(train_df)
```

```python
#Find average relative error

def getPrediction(OverallQual, slope, intercept):
    return slope*OverallQual + intercept

def getRelativeError(prediction, actual):
    return abs(prediction-actual) / actual * 100

def getAccuracy(df, slope, intercept):
    #Gather the average relative errora
    averageRelErrors = []
    for i in range(len(df['SalePrice'])):
        SalePrice = df.iloc[i]['SalePrice']
        OverallQual = df.iloc[i]['OverallQual']
        prediction = getPrediction(OverallQual, slope, intercept)
        relativeError = getRelativeError(prediction, SalePrice)
        #print("Prediction: "+str(prediction)+" Actual: "+str(SalePrice)+" Relative Error: "+str(relativeError)+"%")
        averageRelErrors.append(relativeError)
    return sum(averageRelErrors) / len(averageRelErrors)

```

```python
print("Training Dataset Relative Error: "+str(getAccuracy(train_df, averageSlope, 0))+"%")
```

```python
print("Test Dataset Relative Error: "+str(getAccuracy(test_df, averageSlope, 0))+"%")
```

```python
relativeErrorData = {
    "OverallQual": [],
    "RelativeError": []
}
relativeError_df = pd.DataFrame(relativeErrorData)

# Assuming you have a DataFrame called test_df
for i in range(len(test_df['OverallQual'])):
    OverallQual = test_df.iloc[i]['OverallQual']
    SalePrice = test_df.iloc[i]["SalePrice"]
    prediction = getPrediction(OverallQual, averageSlope, 0)
    relError = getRelativeError(prediction, SalePrice)
    new_row = {'OverallQual': OverallQual, "RelativeError": relError}
    relativeError_df = pd.concat([relativeError_df, pd.DataFrame([new_row])], ignore_index=True)

# Display the resulting DataFrame
print(relativeError_df)
```

```python
# Convert 'OverallQual' to numeric if needed
relativeError_df['OverallQual'] = pd.to_numeric(relativeError_df['OverallQual'], errors='coerce')
relativeError_df.info()
```

```python
relativeError_df.to_csv("relativeError.csv", index=False)
```
